```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("FoodAccessDataOnly.csv")
df.head()
```

```python
df.shape
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
)
```

```python
# basic structure
df.info()
```

```python
# duplicate checks
print("Duplicate rows:", df.duplicated().sum())
print("Unique CensusTract values:", df["censustract"].nunique())
print("Rows:", len(df))
```

```python
# check text columns for whitespace issues
obj_cols = df.select_dtypes(include="object").columns

for col in obj_cols:
    stripped_diff = (df[col].dropna().astype(str) != df[col].dropna().astype(str).str.strip()).sum()
    print(col, "rows with extra whitespace:", stripped_diff)
```

```python
# check for negative numeric values
num_cols = df.select_dtypes(include=[np.number]).columns
negative_counts = (df[num_cols] < 0).sum().sort_values(ascending=False)
negative_counts[negative_counts > 0]
```

```python
# display missing values (without "Low Access" Indicators)
missing_summary = pd.DataFrame({
    "missing_count": df.isna().sum(),
    "missing_percent": 100 * df.isna().mean()
})

missing_summary = missing_summary.loc[~missing_summary.index.str.startswith("la")]
missing_summary = missing_summary.sort_values("missing_count", ascending=False)

missing_summary.head(30)
```
